Generator is quickly then return

```python
def generator():
    print("Start Generator()")
    for x in range(0,5):
        yield x**2

gen = generator()
for i in range(0,5):
    print(next(gen))
```

Dictionary Approach

```python
info = {'a' : 1, 'b' : 2, 'c' : 3, 'd' : 4}

for key, value in info.items():
    print(key, value)
```

Using zip for loop

```python
item1 = [[1,2], [3,4], [5,6]]
item2 = [['a','b'], ['c','d'], ['e','f']]

for digits, characters in zip(item1, item2):
    print(digits, characters)
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)

plt.plot(x,y1,'-o')
plt.plot(x,y2,'--s')
plt.plot(x,y3,':x')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y = x', 'y = 0.25 * x ** 2', 'y = cos(x)'])
plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)

plt.subplot(3,1,1)
plt.plot(x,y1,'-ok')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,2)
plt.plot(x,y2,'--sk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,3)
plt.plot(x,y3,':xk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5., 5., 0.1)
y = np.arange(-5., 5., 0.1)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure(figsize=(5,5))
plt.contourf(x,y,z, cmap = 'coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```
